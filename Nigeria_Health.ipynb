## Exploratory Data Analysis(E.D.A)

## What are the most common causes of death listed in the dataset, and what are their corresponding death rates?

```python
# prompt: Population Current	Population Projected	Cause of Death	Death Rate	Health Statistic Title	Health Statistic Value	Health Target Title	Health Target Value
# 0	0	0	Malaria	83.5	0	0	0	0
# 1	0	0	Lower respiratory infections	83.3	0	0	0	0
# 2	0	0	Diarrhoeal diseases	64.0	0	0	0	0
# 3	0	0	Tuberculosis	55.9	0	0	0	0
# 4	0	0	Maternal conditions	45.4	0	0	0	0

most_common_causes_of_death = df_filled.groupby('Cause of Death')['Death Rate'].mean().sort_values(ascending=False)
print("Most common causes of death and their average death rates:")
most_common_causes_of_death
```

```python
plt.figure(figsize=(10, 6))
most_common_causes_of_death.plot(kind='bar')
plt.title('Average Death Rate by Cause of Death')
plt.xlabel('Cause of Death')
plt.ylabel('Average Death Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## How does the 'Death Rate' vary across different 'Cause of Death'?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (based on your input)
data = {
    'Cause of Death': [
        'Malaria',
        'Lower respiratory infections',
        'Diarrhoeal diseases',
        'Tuberculosis',
        'Maternal conditions'
    ],
    'Death Rate': [83.5, 83.3, 64.0, 55.9, 45.4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Sort data by death rate (descending)
df_sorted = df.sort_values(by='Death Rate', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df_sorted, y='Cause of Death', x='Death Rate', palette='Reds_r', hue='Cause of Death', legend=False)
plt.title('Death Rate by Cause of Death')
plt.xlabel('Death Rate')
plt.ylabel('Cause of Death')
plt.tight_layout()
plt.show()
```

## Are there any outliers in the 'Death Rate' column?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Use the main df_filled DataFrame and filter for rows with Cause of Death
death_rates_data = df_filled[df_filled['Cause of Death'] != 0]['Death Rate']

# Calculate IQR
Q1 = death_rates_data.quantile(0.25)
Q3 = death_rates_data.quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = death_rates_data[(death_rates_data < lower_bound) | (death_rates_data > upper_bound)]

# Display outliers
print("Outliers in 'Death Rate':")
print(outliers if not outliers.empty else "No outliers found.")

# Optional: Boxplot to visualize
plt.figure(figsize=(8, 4))
sns.boxplot(x=death_rates_data, color='tomato')
plt.title('Boxplot of Death Rate')
plt.xlabel('Death Rate')
plt.show()
```

## What is the distribution of 'Death Rate'?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Use the 'Death Rate' column from df_filled where 'Cause of Death' is not 0
death_rates_for_distribution = df_filled[df_filled['Cause of Death'] != 0]['Death Rate']

# Plot histogram + KDE
plt.figure(figsize=(8, 5))
sns.histplot(death_rates_for_distribution, bins=10, kde=True, color='steelblue')
plt.title('Distribution of Death Rate')
plt.xlabel('Death Rate')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Is there any relationship between 'Population Current', 'Population Projected', and the 'Death Rate'?

```python
# Convert to numeric if needed
df_filled['Population Current'] = pd.to_numeric(df_filled['Population Current'], errors='coerce')
df_filled['Population Projected'] = pd.to_numeric(df_filled['Population Projected'], errors='coerce')
df_filled['Death Rate'] = pd.to_numeric(df_filled['Death Rate'], errors='coerce')

# Scatter plots
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df_filled, x='Population Current', y='Death Rate')
plt.title('Death Rate vs Population Current')
plt.show()

sns.scatterplot(data=df_filled, x='Population Projected', y='Death Rate')
plt.title('Death Rate vs Population Projected')
plt.show()

# Correlation check
df_filled[['Population Current', 'Population Projected', 'Death Rate']].corr()
```
